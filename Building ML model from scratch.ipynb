# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Dummy Data Generation

```python
X = np.arange(-4,10,0.3) #we will generate data from -4 to 10 at the unit change of 0.3
Y = -1 * ( X**2 + X -2 )
print("length of the samples of data = ", len(Y))
```

# Data Analysis and Preprocessing

```python
plt.scatter(X,Y ,color = 'r' , label = 'dummy data')
plt.show()
```

```python
#There are no outliers so our machine will easily trend
#So, let us add some of the noise
#  Y = F(X) + eta
# eta has zero mean and sigma square variance
error = 2 #will control the amount of error
Y = -1*(X**2+X-2) + error*np.random.normal(0,3,len(X))
#In random.normal first parameter is the mean and second parameter is the variance
plt.scatter(X,Y ,color = 'r' , label = 'dummy data')
plt.show()
```

## Feature Formation

```python
#   Y = [ X   X^2] * Weight + Bias
#   weight  = [theta1   +   theta2]^T   and Bias = theta0
# Let us make the colmuns in which X . X^2 . Y are the columns
# Let us define the function for that
```

```python
def make_polynomial_feature(X, Y):
    X1 = np.reshape(X , (X.shape[0],1))
    X2 = X**2
    X2 = np.reshape(X2, (X2.shape[0],1))
    new_data = np.concatenate((X1,X2) , axis = 1)
    Y1 = np.reshape(Y, (Y.shape[0],1))
    new_data = np.concatenate((new_data,  Y1), axis =1)
    return new_data
X1 = make_polynomial_feature(X,Y)
X = X1
np.random.shuffle(X) #shuffle along first axis (x axis) to remove increasing pattern of data
```

```python
def prediction(weight, bias , X):
    Y_pred = np.dot(X,weight) + bias
    return Y_pred
```

```python
#now let us write alogirthm for gradient descent
#The learning rate is 0.002 and iter = 999
#some important formula
# error = T_pred - Y_act
# weight = weight - learning_rate * ( 2* error * Xnew)
# bias = bias - learning_rate *2 * sum(error)
def fit_ml(X_train, Y_train, learning_rate = 0.0001, iterations = 1000):
    num_samples , num_features = X_train.shape
    print("the number of samples is ", num_samples)
    print("number of features  = " , num_features)
    weight = np.zeros(num_features)
    bias = 0.1
    for i in range(iterations):
        Y_pred = prediction(weight, bias, X_train)
        dw = (1/num_samples)*2*np.dot(X_train.T , (Y_pred - Y_train))
        db = (1/num_samples)*2*np.sum(Y_pred - Y_train)
        weight = weight - learning_rate*dw
        bias = bias - learning_rate*db
        
    return weight, bias
```

## Split the Data

```python
def make_splitting(data, split_percentage = 0.8):
    X_train = X[:int(split_percentage*len(X)) , :-1]
    Y_train = X[:int(split_percentage*len(X)) ,  -1]
    print(X_train.shape, Y_train.shape)
    X_test = X[int(split_percentage*len(X)) : , : -1]
    Y_test = X[int(split_percentage*len(X)) : ,   -1]
    return X_train, Y_train , X_test, Y_test
X_train, Y_train , X_test, Y_test  = make_splitting(X)
plt.scatter(X_train[:,0], Y_train , color='k' , label = 'Training Data')
plt.scatter(X_test[:,0], Y_test, color='b' , label = 'Testing Data')
plt.xlabel("Samples")
plt.ylabel("Y-axis: Y")
plt.legend()
plt.show()
```

## Evaluation of the model

```python
# we will use MSE to evaluate
def evaluation_mse(Y_pred , Y_act):
    m = len(Y_act)
    mse = 0
    for i in range(m):
        mse = mse + (Y_pred[i] - Y_act[i])**2
    MSE = (1/m)*mse
    return MSE
```

## Training of the Model

```python
weight , bias = fit_ml(X_train, Y_train)
print("Training is completed")
Y_train_pred = prediction(weight, bias , X_train)
print("Mean Squared Error = " , evaluation_mse(Y_train_pred, Y_train))
plt.scatter(X_train[:,0], Y_train , color='k' , label = 'TrainData')
plt.scatter(X_train[:,0], Y_train_pred, color='b' , label = 'prediction Data')
plt.xlabel("Samples")
plt.ylabel("Y-axis: Y")
plt.legend()
plt.grid()
plt.show()
```

## Testing

```python
Y_test_pred = prediction(weight,bias,X_test)
print("Mean Squared Error = " , evaluation_mse(Y_test_pred, Y_test))
plt.scatter(X_test[:,0],Y_test, color = 'r' , label = "test Data")
plt.scatter(X_test[: ,0] , Y_test_pred , color = 'g' , label = " prediction ")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid()
plt.show()
```
